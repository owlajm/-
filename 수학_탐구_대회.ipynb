```python
import sympy as sp

# 이진수 -> 십진수 변환 함수
def binary_to_decimal(binary_str):
    try:
        # 입력값이 올바른 이진수인지 확인
        decimal_value = int(binary_str, 2)
        return decimal_value
    except ValueError:
        return "유효하지 않은 이진수입니다. 0과 1로만 이루어진 숫자를 입력하세요."

# 다항함수 미분 및 미분계수 계산 함수
def differentiate_function():
    x = sp.symbols('x')

    # 다항함수를 입력받아 SymPy 표현식으로 변환
    polynomial_str = input("다항함수를 입력하세요 (예: x**3 + 2*x**2 - 5*x + 4): ")
    try:
        f = sp.sympify(polynomial_str)
    except (sp.SympifyError, ValueError):
        print("올바른 함수 식을 입력하세요.")
        return

    # 미분 계산
    f_prime = sp.diff(f, x)

    # 출력
    print(f"원래 함수: f(x) = {f}")
    print(f"미분한 함수: f'(x) = {f_prime}")

    # x 값 입력 및 도함수 값 계산
    try:
        x_value = float(input("x의 값을 입력하세요: "))
        f_prime_value = f_prime.subs(x, x_value)
        print(f"f'({x_value}) = {f_prime_value}")
    except ValueError:
        print("유효한 숫자를 입력하세요.")
        return

    # 함수와 도함수 그래프 출력
    sp.plot(f, f_prime, (x, -10, 10), legend=True, title="원래 함수와 미분 함수")

# 메인 프로그램
def main():
    print("1. 이진수 -> 십진수 변환")
    print("2. 다차함수 미분 및 미분계수 계산")
    choice = input("선택하세요 (1 또는 2): ")

    if choice == '1':
        binary = input("이진수를 입력하세요: ")
        decimal = binary_to_decimal(binary)
        print(f"십진수: {decimal}")
    elif choice == '2':
        differentiate_function()
    else:
        print("잘못된 입력입니다. 프로그램을 종료합니다.")

# 프로그램 실행
if __name__ == "__main__":
    main()-
```
